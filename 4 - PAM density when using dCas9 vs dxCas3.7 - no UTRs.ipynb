```python
import numpy as np
import pandas as pd
df = pd.read_excel('./source/20190717-IS-noUTRs-df1+2.xlsx', sheetname='Sheet1')
df = df.reset_index()
df[:10]
```

```python
df = df[pd.notnull(df['Intergenic Sequence no UTR'])]
df[:10]
```

```python
df.index
```

```python
str(df.loc[6,"Intergenic Sequence no UTR"])
```

```python
## Make a dataframe with the number of PAM sites of each type for each intergenic sequence

import re
from Bio.Seq import Seq

PAM_types = ["GG", "AGA", "AGC", "AGT", "CGA", "CGC", "CGT", "GGA", "GGC", "GGT", "TGA", "TGC", "TGT", "GAA", "GAT", "CAA"]

df_PAM_types = pd.DataFrame(columns= ["Intergenic Sequence","Gene","Adjacent Gene", "Length"] + PAM_types + ["Non_NGG","dxCas3.7", "dCas9"])

for i in df.index:
    IS = df.loc[i,"Intergenic Sequence no UTR"] 
    if len(IS) > 3:
        PAM_counts = np.zeros(16)
        sequence = str(IS)
        sequence_rc = str((Seq(sequence).back_transcribe().reverse_complement()))
        gene = df.loc[i,"Gene"]
        adjacentGene = df.loc[i,"Adjacent Gene"]
        length = len(sequence)
        
        #print sequence
        #search for each PAM on both strands
        for q in range(len(PAM_types)):
            
            thisCount_top = 0
            thisCount_bot = 0
            for m in re.finditer('(?='+PAM_types[q]+")", sequence): #(?=GG) is to find overlapping matches
                thisCount_top +=1
            for n in re.finditer('(?='+PAM_types[q]+")", sequence_rc): #(?=GG) is to find overlapping matches
                thisCount_bot +=1
            PAM_counts[q] = thisCount_top + thisCount_bot
            #print PAM_types[q]
            #print thisCount_top, thisCount_bot
            
        PAMnonNGG = sum(PAM_counts[1:len(PAM_counts)])
        PAMdxCas37 = sum(PAM_counts)
        PAMdCas9 = PAM_counts[0] + PAM_counts[1] + PAM_counts[4] + PAM_counts[7] + PAM_counts[8] + PAM_counts[9] + PAM_counts[10]
        
        #add to dataframe
        df_PAM_types = df_PAM_types.append({'Intergenic Sequence': sequence, 'Gene': gene, 'Adjacent Gene':adjacentGene, "Length":length, 
                                            'GG':PAM_counts[0], "AGA":PAM_counts[1], "AGC":PAM_counts[2], "AGT":PAM_counts[3], "CGA":PAM_counts[4], "CGC":PAM_counts[5], "CGT":PAM_counts[6], "GGA":PAM_counts[7], "GGC":PAM_counts[8], "GGT":PAM_counts[9], "TGA":PAM_counts[10], "TGC":PAM_counts[11], "TGT":PAM_counts[12], "GAA":PAM_counts[13], "GAT":PAM_counts[14], 
                                            "CAA":PAM_counts[15],"Non_NGG":PAMnonNGG ,"dxCas3.7":PAMdxCas37, "dCas9":PAMdCas9 }, ignore_index=True)
df_PAM_types[:15]
```
